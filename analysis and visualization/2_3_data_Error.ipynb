# Data Error for Gold Control

```python
# Add a new column for the number of errors 

import pandas as pd

# Adicionando a nova coluna com valores zero
df_gold_control['N_er_st_s1']= 0
df_gold_control['N_er_s1_s2']= 0
df_gold_control['N_er_s2_s3']= 0
df_gold_control['N_er_s3_con']= 0
df_gold_control['Total errors']= 0


df_gold_control
```

```python
import pandas as pd

# Definir a função para verificar se a coluna "start" é igual a 0 e atribuir valor nulo às colunas relacionadas aos erros
def verificar_start(row):
    if row['start'] == 0:
        return None
    else:
        return row['N_er_st_s1'], row['N_er_s1_s2'], row['N_er_s2_s3'], row['N_er_s3_con']

# Aplicar a função para cada linha do DataFrame
df_gold_control[['N_er_st_s1', 'N_er_s1_s2', 'N_er_s2_s3', 'N_er_s3_con']] = df_gold_control.apply(verificar_start, axis=1, result_type='expand')
df_gold_control.head(3)
```

```python
# number of errors beteen start and step 1

def calcular_diferenca(row):
    if row['start'] != row['step_1']:
        return abs(row['start'] - row['step_1'])
    else:
        return row['N_er_st_s1']

# Aplicar a função para cada linha do DataFrame
df_gold_control['N_er_st_s1'] = df_gold_control.apply(calcular_diferenca, axis=1)
df_gold_control
```

```python
# number of errors beteen step 1  step 2

def calcular_diferenca(row):
    if row['step_1'] != row['step_2']:
        return abs(row['step_1'] - row['step_2'])
    else:
        return row['N_er_s1_s2']

# Aplicar a função para cada linha do DataFrame
df_gold_control['N_er_s1_s2'] = df_gold_control.apply(calcular_diferenca, axis=1)
df_gold_control
```

```python
# number of errors beteen step 2  step 3

def calcular_diferenca(row):
    if row['step_2'] != row['step_3']:
        return abs(row['step_2'] - row['step_3'])
    else:
        return row['N_er_s2_s3']

# Aplicar a função para cada linha do DataFrame
df_gold_control['N_er_s2_s3'] = df_gold_control.apply(calcular_diferenca, axis=1)
df_gold_control.head(20)
```

```python
# number of errors beteen step 3  confirm

def calcular_diferenca(row):
    if row['step_3'] != row['confirm']:
        return abs(row['step_3'] - row['confirm'])
    else:
        return row['N_er_s3_con']

# Aplicar a função para cada linha do DataFrame
df_gold_control['N_er_s3_con'] = df_gold_control.apply(calcular_diferenca, axis=1)
df_gold_control.head(20)
```

```python
# Definir a função para calcular a soma e atribuir à coluna "Total errors"
def calcular_total_errors(row):
    return row['N_er_st_s1'] + row['N_er_s1_s2'] + row['N_er_s2_s3'] + row['N_er_s3_con']

# Aplicar a função para cada linha do DataFrame
df_gold_control['Total errors'] = df_gold_control.apply(calcular_total_errors, axis=1)
df_gold_control
```

```python
df_gold_control.to_excel('Data_gold_test_error.xlsx', index=False)
```
